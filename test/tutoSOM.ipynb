```python
import IO
import SOM2
import SOMclust
import SOMgraph
```

## Compute the descriptors from the trajectory in dcd format

First create a conf file (makeVectorsFromdcd_PCA.conf) for applications/makeVectorsFromdcd_PCA.py with the following content:

- nframes: the number of frames
- structFile: the pdb of your structure
- projection: set to True
- nProcess: the number of parallel jobs

```python
%%bash
cat makeVectorsFromdcd_PCA.conf
```

Run makeVectorsFromdcd_PCA.py with the corresponding conf file

```python
run makeVectorsFromdcd_PCA.py makeVectorsFromdcd_PCA.conf
```

Load the resulting numpy file (projections.npy) which is the input data for the SOM

```python
inputmatrix = numpy.load('projections.npy')
inputmatrix.shape
```

## Learn the SOM

```python
som = SOM2.SOM(inputmatrix)
```

```python
som.learn(verbose=True)
```

Save the resulting map

```python
numpy.save('smap', som.smap)
```

## Compute the U-matrix and visualize

```python
umat = som.umatrix()
```

```python
matshow(umat)
colorbar()
```

## Compute the Best Matching Units: the position in the map for each frame of the input trajectory

```python
bmus = som.get_allbmus_kdtree()
```

## Flood the U-matrix

```python
clust = SOMclust.clusters(umat, bmus, som.smap)
```

```python
clust.getclusters()
```

Visualization of the flooded U-matrix

```python
imshow(ma.masked_array(clust.umat_cont, clust.mask), interpolation='nearest')
```

## SOM as a graph

Switch to flooded SOM

```python
smap_flood = clust.flood(som.smap)[0]
```

Compute graph

```python
graph = SOMgraph.graph(smap=smap_flood, mask=clust.mask, umat=clust.umat_cont)
```

```python
graph.getgraph()
```

Prune graph

```python
graph.clean_graph()
```

## Projecting data onto the SOM

### The RMSD

Load the trajectory

```python
traj = IO.Trajectory('newtraj78_210521_ca.dcd', struct='2pgb_hydro_ca.pdb')
```

```python
traj.struct.atoms['coord'] = traj.struct.atoms['coord'] - traj.struct.atoms['coord'].mean(axis=0)
traj.align(templatestruct=(traj.struct.atoms['coord']).flatten())
```

```python
i,j = traj.array.shape
print (i,j)
```

```python
traj.array = traj.array.reshape((i, j/3, 3))
traj.array.shape
```

```python
rmsds = sqrt((( traj.array - traj.struct.atoms['coord']*numpy.ones_like(traj.array) )**2).sum(axis=2)).mean(axis=1)
```

```python
histo = hist(rmsds, 100, range=[0,5])
```

```python
X,Y,Z = som.smap.shape
def project(data):
    pmap = zeros((X,Y))
    density = zeros((X,Y))
    for c,e in enumerate(bmus):
        i,j = e
        pmap[i,j]+=data[c]
        density[i,j] += 1
    pmap = pmap / density
    pmap = clust.flood(pmap, clust.x_offset, clust.y_offset, clust.mask)[0]
    return pmap
rmsdmap = project(rmsds)
```

```python
clust.plotclusters(color='k', matrix=rmsdmap, cmap=cm.PRGn, vmax=2.)
```
